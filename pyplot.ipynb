```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import clear_output
import ipywidgets as widgets
```

```python
df = pd.read_csv("Test_Data_2013.csv")
```

```python
# function to get unique values
def unique_dates(array):
    unique = array.unique().tolist()
    unique.sort()
    return unique
```

```python
# %load hours_dict.py

def hours_dict(date):
    day = df[df['date']==date]
    hours = [time.split(':')[0] for time in day['time']]
    hours_dict = {i:hours.count(str(i)) for i in np.arange(24)}
    return hours_dict
```

```python
feb15 = hours_dict('2013_02_15')
plt.plot(feb15.keys(),feb15.values())
plt.show()
```

```python
dropdown_dates = widgets.Dropdown(options=unique_dates(df.date))

plot_output = widgets.Output()

def dropdown_dates_eventhandler(change):
    feb15 = hours_dict(change.new)
    with plot_output:
        plot_output.clear_output()
        plt.plot(feb15.keys(),feb15.values())
        plt.show()

dropdown_dates.observe(dropdown_dates_eventhandler, names='value')
    
display(dropdown_dates)
display(plot_output)
```

```python
feb15 = hours_dict('2013_02_15')
feb16 = hours_dict('2013_02_16')
plt.plot(feb15.keys(),feb15.values())
plt.plot(feb16.keys(),feb16.values())
plt.show()
```

```python
feb15 = hours_dict('2013_02_15')
feb16 = hours_dict('2013_02_16')
plt.plot(feb15.keys(),feb15.values())
plt.plot(feb16.keys(),feb16.values())
plt.xticks(np.arange(0,24,6))
plt.xlim(-3,26)
plt.ylim(-3,28)
plt.xlabel('Hour of the day')
plt.ylabel('Measurements Taken')
plt.title('Measurements By Hour')
plt.show()
```
